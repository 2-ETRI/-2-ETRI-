# lifelog 데이터와 sleep 데이터를 먼저 병합 진행
- 이유: user 데이터는 lifelog 데이터와 sleep 데이터에 모두 붙어야 되기 때문

```python
lifelog_df = pd.read_csv('./2018-2019 data/2018-2019_lifelog.csv')
lifelog_df
```

```python
survey_sleep_df = pd.read_csv('./2018-2019 data/2018-2019_survey_sleep.csv')
survey_sleep_df
```

```python
lifelog_survey_sleep_df=lifelog_df.merge(survey_sleep_df, how = 'outer', on=['userId','date'])
lifelog_survey_sleep_df
```

# lifelog_survey_sleep와 user 개인 정보에 대해 병합

```python
info_df = pd.read_csv('./2018-2019 data/user_info_2019_2018_updated.csv')
info_lifelog_survey_sleep_df = info_df.merge(lifelog_survey_sleep_df, how = 'outer', on=['userId'])
info_lifelog_survey_sleep_df=info_lifelog_survey_sleep_df.sort_values(['userId','date'],axis=0).reset_index(drop=True)
info_lifelog_survey_sleep_df
```

```python
info_lifelog_survey_sleep_df.to_csv('./2018-2019 data/2018-2019_all_data.csv',index=False)
```
