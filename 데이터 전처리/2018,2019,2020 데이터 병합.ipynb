# 2020년도와 2018_2019년도 데이터에서 통합할 수 있는 feature들을 통합한 과정임

```python
df_2020 = pd.read_csv('./2020 data/2020_all_data.csv')

df_2020['actionOption_751'] = df_2020[['actionOption_751','actionOption_752','actionOption_753']].sum(axis=1).astype(int)

df_2020['userId'] = '2020_'+df_2020['userId']

#actionOptionSubOption 정리 완료
df_2020['actionSubOption_1'] = df_2020['actionSubOption_move_method_1']
df_2020['actionSubOption_2'] = df_2020['actionSubOption_move_method_2']
df_2020['actionSubOption_3'] = df_2020[['actionSubOption_move_method_3','actionSubOption_move_method_5',
                                        'actionSubOption_move_method_6']].sum(axis=1)
df_2020['actionSubOption_4'] = df_2020['actionSubOption_move_method_4']
df_2020['actionSubOption_5'] = df_2020['actionSubOption_move_method_7']
df_2020 = df_2020.drop(['actionSubOption_move_method_1','actionSubOption_move_method_2',
                       'actionSubOption_move_method_3','actionSubOption_move_method_4',
                       'actionSubOption_move_method_5','actionSubOption_move_method_6',
                       'actionSubOption_move_method_7'], axis=1)

df_2020['conditionSub1Option_5']=df_2020['conditionSub1Option_0']
df_2020 = df_2020.drop(['conditionSub1Option_0'],axis=1)


df_2020['condition_NOT_ALONE'] =  df_2020['condition_WITH_MANY'] + df_2020['condition_WITH_ONE']
df_2020 = df_2020.drop(['condition_WITH_MANY','condition_WITH_ONE'],axis=1)

df_2020
```

```python
df_2018_2019 = pd.read_csv('./2018-2019 data/2018-2019_all_data.csv')

df_2018_2019['userId'] = '2018_2019_'+df_2018_2019['userId']


#actionOption 정리 완료
df_2018_2019['actionOption_751']=df_2018_2019[['actionOption_0','actionOption_1', 'actionOption_2','actionOption_6']].sum(axis=1).astype(int)
df_2018_2019['actionOption_793']=df_2018_2019[['actionOption_4','actionOption_5']].sum(axis=1).astype(int)
df_2018_2019=df_2018_2019.drop(['actionOption_0','actionOption_1','actionOption_2','actionOption_3',
                               'actionOption_4','actionOption_5','actionOption_6'],axis=1)

#actionOptionSubOption 정리 완료
df_2018_2019['actionSubOption_5']=df_2018_2019['actionSubOption_0']
df_2018_2019 = df_2018_2019.drop(['actionSubOption_0'],axis=1)

#conditionSub1Option 정리 완료
df_2018_2019['conditionSub1Option_5']=df_2018_2019['condition_0']

df_2018_2019['condition_ALONE'] = df_2018_2019['condition_0']
df_2018_2019['condition_NOT_ALONE'] = df_2018_2019['condition_1'] + df_2018_2019['condition_2'] + df_2018_2019['condition_3'] + df_2018_2019['condition_4']
df_2018_2019 = df_2018_2019.drop(['condition_0','condition_1','condition_2','condition_3','condition_4'],axis=1)

df_2018_2019
```

# 2018-2019, 2020년도 데이터를 feature 상관없이 통합한 과정

```python
df = pd.concat([df_2018_2019,df_2020])
```
